```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

The code given in Newman's Exercise 3.1 is written below, tweaking it so that object-oriented plotting was used.

```python
#Code from Exercise 3.1
wavelength = 5.0
k = 2 * math.pi / wavelength
xi0 = 1.0
separation = 20.0                          #Separation of centers in cm
side = 100.0                               #Side of the square in cm
points = 500                               #Number of grid points along each side
spacing = side / points                    #Spacing of points in cm

#Calculates the positions of the center of the circles
x1 = side / 2 + separation / 2
y1 = side / 2
x2 = side / 2 - separation / 2
y2 = side / 2

#Makes an array to store the heights
xi = np.empty([points, points], float)

#Calculates the values in the array
for i in range (points):
    y = spacing * i
    for j in range (points):
        x = spacing * j
        r1 = math.sqrt((x - x1) ** 2 + (y - y1) ** 2)
        r2 = math.sqrt((x - x2) ** 2 + (y - y2) ** 2)
        xi[i, j] = xi0 * math.sin(k * r1) + xi0 * math.sin(k * r2)
        
#Makes the plot
fig, axs = plt.subplots()                   #Initializes the subplot

#Creates the plot in the variable im
im = axs.imshow(xi, origin = "lower", extent = [0, side, 0, side])  

axs.set_title("Density Plot")               #Sets the plot title

axs.set_xlabel("X Position")                #Sets the axes labels
axs.set_ylabel("Y Position")

plt.gray()
plt.show()                                  #Prints the plot
```

The density plot printed above shows the superpositions of two circular sets of sine waves that create and interference pattern with fringes that appear as the gray bars radiating out from the center of the plot.
